```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from datetime import datetime
from dotenv import load_dotenv
import os
import time

# Load environment variables from .env file
load_dotenv()

# =================== CONFIGURATION ===================
EMAIL = os.getenv("EMAIL")
PASSWORD = os.getenv("PASSWORD")
LMS_BASE_URL = os.getenv("LMS_BASE_URL")

# =================== START SELENIUM ===================
driver = webdriver.Chrome()
driver.maximize_window()
driver.get(LMS_BASE_URL)
time.sleep(2)

# =================== LOGIN ===================
driver.find_element(By.ID, "email1").send_keys(USERNAME)
time.sleep(1)
driver.find_element(By.ID, "password1").send_keys(PASSWORD)
time.sleep(1)
driver.find_element(By.XPATH, "//*[@id='__next']/div/div/div/form/button").click()
time.sleep(3)


assignment_list=[
  {
    "title": "FSDU03-MU03_S03: Debouncing & Throttling",
    "schedule": "24-10-2025, 02:00 PM",
    "conclude": "28-10-2025, 11:59 PM",
    "category": "coding",
    "tags": "course_a"
  },
  {
    "title": "FSDU03-MU03_S04: Authentication/Authorization using Firebase",
    "schedule": "06-08-2025, 06:00 PM",
    "conclude": "10-08-2025, 11:59 PM",
    "category": "coding",
    "tags": "course_a"
  },
  {
    "title": "FSDU03-MU03_S05: Linked List - 3",
    "schedule": "07-08-2025, 02:00 PM",
    "conclude": "11-08-2025, 11:59 PM",
    "category": "dsa",
    "tags": "course_b"
  },
  {
    "title": "FSDU03-MU03_S06: Linked List - 4",
    "schedule": "07-08-2025, 06:00 PM",
    "conclude": "11-08-2025, 11:59 PM",
    "category": "dsa",
    "tags": "course_b"
  },
  {
    "title": "FSDU03-MU03_S07: Creating Application using Firebase - 1",
    "schedule": "08-08-2025, 02:00 PM",
    "conclude": "12-08-2025, 11:59 PM",
    "category": "coding",
    "tags": "course_a"
  },
  {
    "title": "FSDU03-MU03_S08: Creating Application using Firebase - 2",
    "schedule": "08-08-2025, 06:00 PM",
    "conclude": "12-08-2025, 11:59 PM",
    "category": "coding",
    "tags": "course_a"
  }
]
```

```python
type = "assignment"
platforms = "Assessment platform"
Assessment_Platform_clients = "Masai LMS"

# =================== HELPERS ===================
def convert_to_datetime_local(val: str) -> str:
    """Converts 'DD-MM-YYYY, hh:mm AM/PM' → 'YYYY-MM-DDTHH:MM'"""
    dt = datetime.strptime(val, "%d-%m-%Y, %I:%M %p")
    return dt.strftime("%Y-%m-%dT%H:%M")

def react_input(driver, element, value):
    """Triggers real React-compatible input event"""
    driver.execute_script("""
        const input = arguments[0];
        const value = arguments[1];
        const lastValue = input.value;
        input.value = value;
        const event = new Event('input', { bubbles: true });
        event.simulated = true;
        const tracker = input._valueTracker;
        if (tracker) { tracker.setValue(lastValue); }
        input.dispatchEvent(event);
    """, element, value)

# ==== MAIN LOOP ====
for assignment in assignment_list:
    driver.get("https://experience-admin.masaischool.com/assignment/?page=0")
    time.sleep(3)
    
    title = assignment["title"]
    category = assignment["category"]
    tags = assignment["tags"]
    schedule_ISO_value = convert_to_datetime_local(assignment["schedule"])
    conclude_ISO_value = convert_to_datetime_local(assignment["conclude"])
    
    section = title.split("_", 1)[0].strip()
    batch = "FSD" if section.startswith("F") else "PSD"
    
    # Open Create Assignment
    driver.find_element(By.XPATH, "/html/body/div/div/div/main/div[1]/div[2]/a[2]/button").click()
    time.sleep(2)
    
    # Title
    title_input = driver.find_element(By.XPATH, "/html/body/div/div/div/main/form/label/div/div/input")
    react_input(driver, title_input, title)
    time.sleep(1)
    
    # Assignment classification
    type_input = driver.find_element(By.XPATH, "/html/body/div/div/div/main/form/div[2]/div[1]/div/label[1]/div/div/div[1]/div[2]/input")
    react_input(driver, type_input, type)
    type_input.send_keys(Keys.ENTER)
    time.sleep(1)
    
    category_input = driver.find_element(By.XPATH, "/html/body/div/div/div/main/form/div[2]/div[1]/div/label[2]/div/div/div[1]/div[2]/input")
    react_input(driver, category_input, category)
    category_input.send_keys(Keys.ENTER)
    time.sleep(1)
    
    tags_input = driver.find_element(By.XPATH, "/html/body/div/div/div/main/form/div[2]/div[1]/div/label[3]/div/div/div[1]/div/input")
    react_input(driver, tags_input, tags)
    tags_input.send_keys(Keys.ENTER)
    time.sleep(1)
    
    platforms_input = driver.find_element(By.XPATH, "/html/body/div/div/div/main/form/div[2]/div[1]/div/label[4]/div/div/div[1]/div[2]/input")
    react_input(driver, platforms_input, platforms)
    platforms_input.send_keys(Keys.ENTER)
    time.sleep(1)
    
    Assessment_Platform_clients_input = driver.find_element(By.XPATH, "/html/body/div/div/div/main/form/div[2]/div[1]/div/label[5]/div/div/div[1]/div[2]/input")
    react_input(driver, Assessment_Platform_clients_input, Assessment_Platform_clients)
    Assessment_Platform_clients_input.send_keys(Keys.ENTER)
    time.sleep(1)
    
    # Select Assessment Template
    driver.find_element(By.XPATH, "/html/body/div/div/div/main/form/div[2]/div[1]/div/label[6]/div/div/div").click()
    time.sleep(1)
    search_box = driver.find_element(By.XPATH, "/html/body/div/div/div/main/div/div/div/div[2]/div/div/div[2]/div[1]/div/input")
    react_input(driver, search_box, title)
    time.sleep(1)
    driver.find_element(By.XPATH, "/html/body/div/div/div/main/div/div/div/div[2]/div/div/table/tbody/tr[1]/td[1]/input").click()
    time.sleep(1)
    
    # Batch & Section (React-compatible)
    batch_input = driver.find_element(
        By.XPATH,
        "/html/body/div/div/div/main/form/div[2]/div[2]/div/label[1]/div/div/div[1]/div[2]/input"
    )

    # ✅ Set value via React synthetic event
    driver.execute_script("""
      const input = arguments[0];
      const lastValue = input.value;
      input.value = arguments[1];
      const event = new Event('input', { bubbles: true });
      event.simulated = true;
      const tracker = input._valueTracker;
      if (tracker) tracker.setValue(lastValue);
      input.dispatchEvent(event);
    """, batch_input, batch)

    time.sleep(0.5)
    batch_input.send_keys(Keys.ENTER)
    time.sleep(1)

    # Section Input
    section_input = driver.find_element(
        By.XPATH,
        "/html/body/div/div/div/main/form/div[2]/div[2]/div/label[2]/div/div/div[1]/div[2]/input"
    )

    driver.execute_script("""
      const input = arguments[0];
      const lastValue = input.value;
      input.value = arguments[1];
      const event = new Event('input', { bubbles: true });
      event.simulated = true;
      const tracker = input._valueTracker;
      if (tracker) tracker.setValue(lastValue);
      input.dispatchEvent(event);
    """, section_input, section)

    time.sleep(0.5)
    section_input.send_keys(Keys.ENTER)
    time.sleep(1)

    
    associated_lecture_input = driver.find_element(By.XPATH, "/html/body/div/div/div/main/form/div[2]/div[2]/div/label[3]/div/div/div[1]/div[2]/input")
    react_input(driver, associated_lecture_input, title)
    associated_lecture_input.send_keys(Keys.ENTER)
    time.sleep(1)
    
    # Learning objective info
    group_type_input = driver.find_element(By.XPATH, "/html/body/div/div/div/main/form/div[2]/div[3]/div/div/label[1]/div/div/div[1]/div[2]/input")
    react_input(driver, group_type_input, "Test Group")
    group_type_input.send_keys(Keys.ENTER)
    time.sleep(1)
    
    topic_input = driver.find_element(By.XPATH, "/html/body/div/div/div/main/form/div[2]/div[3]/div/div/label[2]/div/div/div[1]/div[2]/input")
    react_input(driver, topic_input, "topic_title_002")
    topic_input.send_keys(Keys.ENTER)
    time.sleep(1)
    
    learning_objective_input = driver.find_element(By.XPATH, "/html/body/div/div/div/main/form/div[2]/div[3]/div/div/label[3]/div/div/div[1]/div[2]/input")
    react_input(driver, learning_objective_input, "test_LO_003")
    learning_objective_input.send_keys(Keys.ENTER)
    time.sleep(1)
    
    # Schedule (React aware)
    schedule_input = driver.find_element(By.XPATH, "/html/body/div/div/div/main/form/div[2]/div[4]/div/label[1]/div/div/input")
    conclude_input = driver.find_element(By.XPATH, "/html/body/div/div/div/main/form/div[2]/div[4]/div/label[2]/div/div/input")
    react_input(driver, schedule_input, schedule_ISO_value)
    time.sleep(1)
    react_input(driver, conclude_input, conclude_ISO_value)
    time.sleep(1)
    
    # Show score checkbox
    show_score = driver.find_element(By.XPATH, "/html/body/div/div/div/main/form/div[3]/label[1]/label/input")
    driver.execute_script("arguments[0].click();", show_score)
    time.sleep(1)
    
    # Scroll and Click Create button
    create_button = driver.find_element(By.XPATH, "/html/body/div/div/div/main/form/div[1]/div/button")
    driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", create_button)
    time.sleep(1)
    driver.execute_script("arguments[0].click();", create_button)
    time.sleep(2)
    
    print(f"✅ Assignment '{title}' created successfully!\n")

driver.quit()
```
