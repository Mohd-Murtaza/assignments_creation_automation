```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from dotenv import load_dotenv
import os
import time

# Load environment variables from .env file
load_dotenv()

# ==== CONFIGURATION ====
EMAIL = os.getenv("ASSESS_EMAIL")
PASSWORD = os.getenv("ASSESS_PASSWORD")
LMS_BASE_URL = os.getenv("ASSESS_LMS_BASE_URL")

# ==== START SELENIUM ====
driver = webdriver.Chrome()
driver.maximize_window()

driver.get(LMS_BASE_URL)

#Login 
time.sleep(2)
driver.find_element(By.XPATH, "/html/body/div[1]/div/div[1]/main/div/div/div[1]/form/div[1]/input").send_keys(USERNAME)
time.sleep(2)
driver.find_element(By.XPATH, "/html/body/div[1]/div/div[1]/main/div/div/div[1]/form/div[2]/input").send_keys(PASSWORD)
time.sleep(2)
driver.find_element(By.ID, "submitBtn").click()
time.sleep(2)
driver.find_element(By.XPATH, "/html/body/div[3]/div[3]/div/section/div/div/div/div/select/option[3]").click()
time.sleep(2)
driver.find_element(By.XPATH, "/html/body/div[1]/div/div[1]/div/div/div[1]/a[1]").click()
time.sleep(2)
assignment_list=[
  "FSDU05-MU03_S05: Introduction to Dynamic Programming",
  "FSDU05-MU03_S06: PSC on DP - 2",
  "FSDU05-MU03_S07: Authentication",
  "FSDU05-MU03_S08: Authorization"
]
```

```python
for assignment in assignment_list:
    section, title= assignment.split(": ", 1)
    batch_name="PSD"
    if section.startswith("F"):
        batch_name = "FSD"
    print(f"Processing {section} - {title}")
    driver.find_element(By.XPATH, "/html/body/div[1]/div/div[1]/main/div/div/div/div/div[2]/div/input[1]").send_keys(title)
    time.sleep(2)
    driver.find_element(By.XPATH, "/html/body/div[1]/div/div[1]/main/div/div/div/div/div[3]/table/tbody/tr[1]/td[2]/button").click()
    time.sleep(2)
    driver.find_element(By.XPATH, "/html/body/div[4]/div/div/div/div[2]/div[1]/div/label[2]/span[1]").click()
    time.sleep(2)
    driver.find_element(By.XPATH, "/html/body/div[4]/div/div/div/div[2]/div[2]/button[2]").click()
    time.sleep(2)
    print(f"clone done for {section}{title}")
    driver.find_element(By.XPATH, "/html/body/div[1]/div/div[1]/main/div/div/div/div/div[3]/table/tbody/tr[1]/td[3]/a").click()
    time.sleep(2)
    driver.find_element(By.XPATH, "/html/body/div[1]/div/div[1]/main/div/div/div/div/div[1]/div/div[2]/a/button").click()
    time.sleep(2)
    assignment_title_input = driver.find_element(By.XPATH, "/html/body/div[1]/div/div[1]/main/div/div/div/div/div[2]/div[1]/div[2]/div[1]/input")
    assignment_title_input.send_keys(Keys.COMMAND + "a")
    assignment_title_input.send_keys(Keys.DELETE)
    time.sleep(2)
    assignment_title_input.send_keys(assignment)
    time.sleep(2)
    assignment_label_input = driver.find_element(By.XPATH, "/html/body/div[1]/div/div[1]/main/div/div/div/div/div[2]/div[1]/div[2]/div[2]/div/div/input")
    assignment_label_input.send_keys(Keys.COMMAND + "a")
    assignment_label_input.send_keys(Keys.DELETE)
    time.sleep(2)
    print(assignment, "<====== before the assignment_label_input")
    assignment_label_input.send_keys(assignment)
    time.sleep(2)
    batch_name_input = driver.find_element(By.XPATH, "/html/body/div[1]/div/div[1]/main/div/div/div/div/div[2]/div[1]/div[2]/div[3]/div/div/input")
    batch_name_input.send_keys(Keys.COMMAND + "a")
    batch_name_input.send_keys(Keys.DELETE)
    time.sleep(2)
    batch_name_input.send_keys(batch_name)
    time.sleep(2)
    driver.find_element(By.XPATH, "/html/body/div[1]/div/div[1]/main/div/div/div/div/div[2]/div[2]/button").click()
    time.sleep(2)
    print(f"Assignment=> {assignment} clone successfully")
```
